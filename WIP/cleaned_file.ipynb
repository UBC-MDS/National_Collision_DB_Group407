```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("../data/raw_data.csv")
```

```python
X = df.drop(["C_YEAR", "C_VEHS", "C_CONF", "V_ID",
             "P_ID", "P_PSN", "P_ISEV", "P_USER", 
             "C_CASE", "C_SEV", "P_SEX", "P_SAFE", "P_AGE", "V_YEAR"], axis=1)

y = df["C_SEV"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=407)
combined_train = pd.concat([X_train, y_train], axis=1)
```

```python
drop_strings = ((combined_train.C_MNTH != "UU")& 
         (combined_train.C_WDAY != "U")  & 
         (combined_train.C_HOUR != "UU") & 
         (combined_train.C_RCFG != "UU") & 
         (combined_train.C_RCFG != "QQ") & 
         (combined_train.C_WTHR != "U") & 
         (combined_train.C_WTHR != "Q") &
         (combined_train.C_RSUR != "Q") &
         (combined_train.C_RSUR != "U") &
         (combined_train.C_RALN != "Q") &
         (combined_train.C_RALN != "U") &
         (combined_train.C_TRAF != "QQ") &
         (combined_train.C_TRAF != "UU") &
         (combined_train.V_TYPE != "NN") & 
         (combined_train.V_TYPE != "UU") & 
         (combined_train.V_TYPE != "QQ"))
```

```python
clean = combined_train[drop_strings]
clean.apply(pd.to_numeric);
```

```python
final_clean = pd.get_dummies(clean)
```

```python
df = final_clean[set(final_clean.columns)]
```
